```python
import getpass
import os

os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_ENDPOINT"] = "https://api.smith.langchain.com"
os.environ["LANGCHAIN_PROJECT"] = "pr-langchain-test"
os.environ["LANGCHAIN_API_KEY"] = getpass.getpass()
```

```python
# @title Load the pdf document {"vertical-output":true,"display-mode":"form"}
from langchain_community.document_loaders import PyPDFLoader

file_path = "./drive/MyDrive/sample_data_rag/nke-10k-2023.pdf"
loader = PyPDFLoader(file_path)

docs = loader.load()

print(len(docs))
print(docs[1].metadata)
#docs = docs[0:3]
```

```python
# @title Split document into chunks
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000, chunk_overlap=200, add_start_index=True
)
all_splits = text_splitter.split_documents(docs)

print(f"Split blog post into {len(all_splits)} sub-documents.")
```

```python
# @title Embeddings
from google.colab import userdata

from langchain_huggingface import HuggingFaceEmbeddings


hf_token = userdata.get('HF_TOKEN')
print(hf_token)

model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {'device': 'cpu'}
encode_kwargs = {'normalize_embeddings': False}
embeddings = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)
```

```python
!pip install -qU langchain-chroma
```

```python
# @title Use embeddings to create Vecotr and Store it in vector database
from langchain_chroma import Chroma

vector_store = Chroma(embedding_function=embeddings)
ids = vector_store.add_documents(documents=all_splits)

print(ids)

```

```python
# @title Return documents based on similarity to a string query
results = vector_store.similarity_search(
    "How many distribution centers does Nike have in the US?"
)

print(results[0])
```

```python
# @title Return documents based on similarity to an embedded query
embedding = embeddings.embed_query("How many distribution centers does Nike have in the US?")
#embedding = embeddings.embed_query("How were Nike's margins impacted in 2023?")

print(embedding)

results = vector_store.similarity_search_by_vector(embedding)
print(results[0])
```
